### One-Hot Encoding
Преобразуем категориальные признаки в числовой формат с помощью One-Hot Encoding. Каждая категория становится отдельной бинарной колонкой. Параметр `drop_first=True` удаляет первую категорию для избежания мультиколлинеарности (дамми-ловушка).

```python
df_encoded = df_clean.copy()
df_encoded = pd.get_dummies(df_encoded, columns=cat_cols, drop_first=True)

print("ДО кодирования (первые 5 строк):")
display(df_clean[cat_cols].head())

print("ПОСЛЕ кодирования (все новые колонки, первые 5 строк):")
new_cols = [col for col in df_encoded.columns if any(cat in col for cat in cat_cols)]
display(df_encoded[new_cols].head())
```

**Важно:** One-Hot Encoding эффективен когда у категориальных признаков немного уникальных значений (обычно < 15). Если значений много, датафрейм становится слишком разряженным и разрастается в сотни колонок.


---


### Label Encoding
Для порядковых категорий (где важен порядок) используем Label Encoding - заменяем категории числами 0, 1, 2... Важно: подходит только если категории имеют естественный порядок.

**Для наших данных:** Label Encoding нецелесообразен, так как все наши категориальные признаки номинативные (без порядка). Исключение - колонки OverallQual и OverallCond, но они уже представлены числами.


---





### Feature Hashing
Для признаков с большим количеством уникальных значений (например, район) используем Feature Hashing. Сжимаем множество категорий в заданное число признаков (n_features=10), экономя память.

```python
from sklearn.feature_extraction import FeatureHasher

high_card_cols = []
for col in cat_cols:
    unique_count = df_clean[col].nunique()
    print(f"{col}: {unique_count} уникальных значений")
    if unique_count > 13:
        high_card_cols.append(col)

print("\nКолонки для хеширования:", high_card_cols)

if high_card_cols:
    for col in high_card_cols:
        hash_input = df_clean[col].apply(lambda x: {col: x}).tolist()
        hasher = FeatureHasher(n_features=10, input_type='dict')
        hashed = hasher.transform(hash_input).toarray()
        
        hashed_df = pd.DataFrame(hashed, columns=[f'{col}_hash_{i}' for i in range(10)])
        print(f"\n{col}: {df_clean[col].nunique()} уникальных -> 10 хеш-признаков")
        display(hashed_df.head())
else:
    print("Нет колонок с >13 уникальными значениями")
```
