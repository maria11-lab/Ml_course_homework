<div style="padding: 20px; border-radius: 12px; border-left: 5px solid #C33F4C; margin: 15px 0; background: rgba(195, 63, 76, 0.1); backdrop-filter: blur(5px); box-shadow: 0 5px 15px rgba(0,0,0,0.08); border: 1px solid rgba(195, 63, 76, 0.2);"> 
 <p style="font-size: 15px; line-height: 1.5; color: #2c3e50; margin: 0;"> 
   <b style="color: #C33F4C;">Пропуски:</b> Считаем количество и процент. PoolQC (99.5%), Alley (93.8%), MiscFeature (96.3%) — почти пустые. Возможно, эти признаки не так важны, и мы можем убрать их при анализе данных.
 </p>
</div>

```python
nulls = df.isnull().sum()
percent_missing = (df.isnull().mean() * 100).round(2)

missing_df = pd.DataFrame({
    'Количество пропусков': nulls,
    'Процент пропусков': percent_missing
})

missing_df = missing_df[missing_df['Количество пропусков'] > 0]

print("Анализ пропусков в данных:")
print(missing_df)
```

```python
print(f"Количество полных дубликатов: {df.duplicated().sum()}")
```

<div style="padding: 20px; border-radius: 12px; border-left: 5px solid #C33F4C; margin: 15px 0; background: rgba(195, 63, 76, 0.1); backdrop-filter: blur(5px); box-shadow: 0 5px 15px rgba(0,0,0,0.08); border: 1px solid rgba(195, 63, 76, 0.2);"> 
 <p style="font-size: 15px; line-height: 1.5; color: #2c3e50; margin: 0;"> 
   <b style="color: #C33F4C;">Итог:</b> Дубликатов нет.
 </p>
</div> 

<div style="padding: 20px; border-radius: 12px; border-left: 5px solid #C33F4C; margin: 15px 0; background: rgba(195, 63, 76, 0.1); backdrop-filter: blur(5px); box-shadow: 0 5px 15px rgba(0,0,0,0.08); border: 1px solid rgba(195, 63, 76, 0.2);"> 
 <p style="font-size: 15px; line-height: 1.5; color: #2c3e50; margin: 0;"> 
   <b style="color: #C33F4C;">Удаление строк с пропусками (dropna):</b> В столбце Electrical всего 1 пропуск. Удаляем эту строку, это не повлияет на анализ.
 </p>
</div>

```python
df_clean = df.copy()

df_clean = df_clean.dropna(subset=['Electrical'])
print(f"Строк после удаления пропуска в Electrical: {len(df_clean)}")
```

<div style="padding: 20px; border-radius: 12px; border-left: 5px solid #C33F4C; margin: 15px 0; background: rgba(195, 63, 76, 0.1); backdrop-filter: blur(5px); box-shadow: 0 5px 15px rgba(0,0,0,0.08); border: 1px solid rgba(195, 63, 76, 0.2);"> 
 <p style="font-size: 15px; line-height: 1.5; color: #2c3e50; margin: 0;"> 
   <b style="color: #C33F4C;">Удаление бесполезных столбцов:</b> Столбцы с >80% пропусков не несут информации. Удаляем их.
 </p>
</div>

```python
threshold = 80

cols_to_drop = missing_df[missing_df['Процент пропусков'] > threshold].index.tolist()
print(f"Столбцы с >{threshold}% пропусков (удаляем):")
print(cols_to_drop)

df_clean = df_clean.drop(columns=cols_to_drop)
```

<div style="padding: 20px; border-radius: 12px; border-left: 5px solid #C33F4C; margin: 15px 0; background: rgba(195, 63, 76, 0.1); backdrop-filter: blur(5px); box-shadow: 0 5px 15px rgba(0,0,0,0.08); border: 1px solid rgba(195, 63, 76, 0.2);"> 
 <p style="font-size: 15px; line-height: 1.5; color: #2c3e50; margin: 0;"> 
   <b style="color: #C33F4C;">Заполнение средним (mean):</b> LotFrontage имеет нормальное распределение. Заполняем пропуски средним.
 </p>
</div>

```python
mean_value = df_clean['LotFrontage'].mean()
df_clean['LotFrontage'] = df_clean['LotFrontage'].fillna(mean_value)
print(f"LotFrontage заполнен средним: {mean_value:.2f}")
```

<div style="padding: 20px; border-radius: 12px; border-left: 5px solid #C33F4C; margin: 15px 0; background: rgba(195, 63, 76, 0.1); backdrop-filter: blur(5px); box-shadow: 0 5px 15px rgba(0,0,0,0.08); border: 1px solid rgba(195, 63, 76, 0.2);"> 
 <p style="font-size: 15px; line-height: 1.5; color: #2c3e50; margin: 0;"> 
   <b style="color: #C33F4C;">Заполнение медианой (median):</b> LotArea содержит выбросы. Медиана лучше подходит для заполнения.
 </p>
</div>

```python
median_value = df_clean['LotArea'].median()
df_clean['LotArea'] = df_clean['LotArea'].fillna(median_value)
print(f"LotArea заполнен медианой: {median_value:.2f}")

# MasVnrArea (площадь облицовки) - заполняем медианой из-за огромного разброса значений
# Медиана = 0, что логично, т.к. у многих домов нет облицовки
df_clean['MasVnrArea'] = df_clean['MasVnrArea'].fillna(df_clean['MasVnrArea'].median())
print(f"MasVnrArea заполнен медианой: {df_clean['MasVnrArea'].median()}")
```

<div style="padding: 20px; border-radius: 12px; border-left: 5px solid #C33F4C; margin: 15px 0; background: rgba(195, 63, 76, 0.1); backdrop-filter: blur(5px); box-shadow: 0 5px 15px rgba(0,0,0,0.08); border: 1px solid rgba(195, 63, 76, 0.2);"> 
 <p style="font-size: 15px; line-height: 1.5; color: #2c3e50; margin: 0;"> 
   <b style="color: #C33F4C;">Заполнение константой:</b> В FireplaceQu пропуски означают отсутствие камина. Заменяем на 'NoFireplace'.
 </p>
</div>

```python
df_clean['FireplaceQu'] = df_clean['FireplaceQu'].fillna('NoFireplace')
print("FireplaceQu заполнен константой 'NoFireplace'")
print(f"Пропусков после заполнения: {df_clean['FireplaceQu'].isnull().sum()}")
```

<div style="padding: 20px; border-radius: 12px; border-left: 5px solid #C33F4C; margin: 15px 0; background: rgba(195, 63, 76, 0.1); backdrop-filter: blur(5px); box-shadow: 0 5px 15px rgba(0,0,0,0.08); border: 1px solid rgba(195, 63, 76, 0.2);"> 
 <p style="font-size: 15px; line-height: 1.5; color: #2c3e50; margin: 0;"> 
   <b style="color: #C33F4C;">Заполнение модой (mode):</b> Для всех категориальных признаков находим самое частое значение и заполняем пропуски.
 </p>
</div>

```python
cat_cols = df_clean.select_dtypes(include=['str']).columns
cat_cols_with_nulls = [col for col in cat_cols if df_clean[col].isnull().any()]

for col in cat_cols_with_nulls:
    mode_val = df_clean[col].mode()[0]
    df_clean[col] = df_clean[col].fillna(mode_val)
    print(f"{col} заполнен модой: '{mode_val}'")
```

<div style="padding: 20px; border-radius: 12px; border-left: 5px solid #C33F4C; margin: 15px 0; background: rgba(195, 63, 76, 0.1); backdrop-filter: blur(5px); box-shadow: 0 5px 15px rgba(0,0,0,0.08); border: 1px solid rgba(195, 63, 76, 0.2);"> 
 <p style="font-size: 15px; line-height: 1.5; color: #2c3e50; margin: 0;"> 
   <b style="color: #C33F4C;">Заполнение года постройки гаража:</b> Для 81 дома с пропусками мы уже заполнили категориальные признаки модой. Год постройки гаража логично взять равным году постройки дома.
 </p>
</div>

```python
df_clean['GarageYrBlt'] = df_clean['GarageYrBlt'].fillna(df_clean['YearBuilt'])
print("GarageYrBlt заполнен годом постройки дома")
```

<div style="padding: 20px; border-radius: 12px; border-left: 5px solid #C33F4C; margin: 15px 0; background: rgba(195, 63, 76, 0.1); backdrop-filter: blur(5px); box-shadow: 0 5px 15px rgba(0,0,0,0.08); border: 1px solid rgba(195, 63, 76, 0.2);"> 
 <p style="font-size: 15px; line-height: 1.5; color: #2c3e50; margin: 0;"> 
   <b style="color: #C33F4C;">Базовая статистика числовых признаков:</b> Смотрим минимум, максимум, среднее, медиану и моду.
 </p>
</div>

```python
numeric_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns

basic_stats = pd.DataFrame({
    'min': df_clean[numeric_cols].min(),
    'max': df_clean[numeric_cols].max(),
    'mean': df_clean[numeric_cols].mean(),
    'median': df_clean[numeric_cols].median(),
    'mode': df_clean[numeric_cols].mode().iloc[0]
})

print("Базовая статистика (min, max, mean, median, mode):")
basic_stats.round(2)
```

<div style="padding: 20px; border-radius: 12px; border-left: 5px solid #C33F4C; margin: 15px 0; background: rgba(195, 63, 76, 0.1); backdrop-filter: blur(5px); box-shadow: 0 5px 15px rgba(0,0,0,0.08); border: 1px solid rgba(195, 63, 76, 0.2);"> 
 <p style="font-size: 15px; line-height: 1.5; color: #2c3e50; margin: 0;"> 
   <b style="color: #C33F4C;">Перцентили (квантили):</b> 5%, 25%, 50%, 75%, 95% — показывают распределение значений.
 </p>
</div>

```python
percentiles = pd.DataFrame({
    '5%': df_clean[numeric_cols].quantile(0.05),
    '25%': df_clean[numeric_cols].quantile(0.25),
    '50%': df_clean[numeric_cols].quantile(0.5),
    '75%': df_clean[numeric_cols].quantile(0.75),
    '95%': df_clean[numeric_cols].quantile(0.95)
})

print("Перцентили (5%, 25%, 50%, 75%, 95%):")
percentiles.round(2)
```

<div style="padding: 20px; border-radius: 12px; border-left: 5px solid #C33F4C; margin: 15px 0; background: rgba(195, 63, 76, 0.1); backdrop-filter: blur(5px); box-shadow: 0 5px 15px rgba(0,0,0,0.08); border: 1px solid rgba(195, 63, 76, 0.2);"> 
 <p style="font-size: 15px; line-height: 1.5; color: #2c3e50; margin: 0;"> 
   <b style="color: #C33F4C;">Продвинутая статистика:</b><br>
   • Дисперсия — разброс значений<br>
   • Асимметрия — перекос распределения (>0 — хвост справа, есть очень большие значения)<br>
   • Эксцесс — острота пика (>0 — много выбросов)
 </p>
</div>

```python

advanced_stats = pd.DataFrame({
    'variance': df_clean[numeric_cols].var(),
    'skewness': df_clean[numeric_cols].skew(),
    'kurtosis': df_clean[numeric_cols].kurtosis()
})

advanced_stats.round(2)

print('-'*50)
print("ИНТЕРПРЕТАЦИЯ ДЛЯ SalePrice:")
print('-'*50)

print(f"SalePrice variance: {df_clean['SalePrice'].var():.2f} (большой разброс цен)")
print(f"SalePrice skewness: {df_clean['SalePrice'].skew():.2f} — {'правый хвост (есть очень дорогие дома)' if df_clean['SalePrice'].skew() > 0.5 else 'симметричное'}")
print(f"SalePrice kurtosis: {df_clean['SalePrice'].kurtosis():.2f} — {'много выбросов' if df_clean['SalePrice'].kurtosis() > 3 else 'умеренный эксцесс'}")
```
